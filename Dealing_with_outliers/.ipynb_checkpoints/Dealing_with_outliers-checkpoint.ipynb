## Dealing with outliers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Ames_Housing_Data.csv")
df.head()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
```

```python
df.select_dtypes(include=['number']).corr()['SalePrice'].sort_values()
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
sns.distplot(df["SalePrice"],kde=True)
```

```python
sns.scatterplot(x='Overall Qual',y='SalePrice',data=df)
```

```python
df[(df['Overall Qual']>8) & (df['SalePrice']<200000)]
```

```python
sns.scatterplot(x='Gr Liv Area',y='SalePrice',data=df)
```

```python
numeric_df[(numeric_df['Gr Liv Area'] > 4000) & (numeric_df['SalePrice'] < 400000)]
```

```python
numeric_df[(numeric_df['Gr Liv Area']>4000) & (numeric_df['SalePrice']<400000)].index
```

```python
ind_drop = df[(df['Gr Liv Area']>4000) & (df['SalePrice']<400000)].index
```

```python
df = df.drop(ind_drop,axis=0)
```

```python
sns.scatterplot(x='Gr Liv Area',y='SalePrice',data=df)
```

```python
df.to_csv("Ames_outliers_removed.csv",index=False)
```
